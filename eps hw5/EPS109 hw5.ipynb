```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#all subparts of part 1 were repeated with these cells
```

```python
def mandelbrot(z,C):
    zNew = z**2 + C
    return zNew
```

```python
def f(z_naught,nmax,zlim):
    z=z_naught
    for i in range(nmax):
        z = mandelbrot(z,C)
        if np.abs(z)>=zlim:
            return i
    return nmax
```

```python
C = complex(-0.85,0.22)
N=501
step = (3/N)
Re = np.arange(-1.5,1.5,step)
Im = np.arange(-1.5,1.5,step)
grid = np.zeros((N,N))

for i in range(N):
    for j in range(N):
        z0 = complex(Re[i],Im[j])
        x = f(z0,100,100)
        grid[i,j]=x
```

```python
plt.figure(figsize=(9.6,7.2))
plt.imshow(grid,cmap='inferno')

#plt.title(r'First Fractal with $z_0$ = [-1.5,1.5]$x$ + [-1.5,1.5]$iy$')
#plt.title(r'First Fractal zoom with $z_0$ = [0.15,0.151]x + [-0.001,0]$iy$')
plt.title(r'Second Fractal with $z_0$ = [-1.5,1.5]$x$ + [-1.5,1.5]$iy$')
#plt.title(r'Fractal with C = 0.4 - 0.2$i$ and $z_0$ = [-0.245,-0.244]x + [-0.032,-0.031]$iy$')

#plt.savefig('First Fractal.jpeg')
#plt.savefig('First Fractal zoom.jpeg')
plt.savefig('Second Fractal.jpeg')
#plt.savefig('Explored Fractal.jpeg')
plt.show()
```
