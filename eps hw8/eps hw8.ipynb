## (1)

```python
N = 50
d = 5
L = (4*d) + d
T = np.zeros(5*N)

step = 0.1
x = np.arange(0,L,step)
```

```python
def T_naught(T):
    for i in range(T.shape[0]):
        if x[i]>10 and x[i]<=15:
            T[i]=1423 # 1150 C in Kelvin
        else:
            T[i]=273 # 0 Deg C in Kelvin
    return T
```

```python
T = T_naught(T)
plt.plot(x,T)
plt.ylim(270,1500)
plt.show()
```

```python
c = 1020 # [J/(kg*K)]
rho = 2850 # [kg/(m^3)]
k = 2.8 # [J/(m*s*K)]
dt = 1800 # [s]
dx_sq = (L/(N*5))**2 # [m^2]
n = (dt/dx_sq)*(k/(c*rho))

def fn(T,n):
    Tnew=np.copy(T)
    for i in range(1,N*5-1):
        Tnew[i] = (n*((T[i+1])-(2*T[i])+T[i-1]))+(T[i])
    return Tnew
```

```python
T = T_naught(T)
for j in range(15000):
    T=fn(T,n)
    
plt.plot(x,T,'ro-',mfc='w')
plt.ylim(273,1500)
plt.show()
```

```python
T = T_naught(T)
for j in range(60*60//1800):
    T=fn(T,n)
    
plt.plot(x,T,'ro-',mfc='w')
plt.ylim(273,1500)
plt.show()

Tdike = T[(T.size-1)//2]
print('Dike temp after 1 hour',Tdike,'[K]')
```

```python
T = T_naught(T)
for j in range(24*60*60//1800):
    T=fn(T,n)
    
plt.plot(x,T,'ro-',mfc='w')
plt.ylim(273,1500)
plt.show()

Tdike = T[(T.size-1)//2]
print('Dike temp after 1 day',Tdike,'[K]')
```

```python
T = T_naught(T)
for j in range(7*24*60*60//1800):
    T=fn(T,n)
    
plt.plot(x,T,'ro-',mfc='w')
plt.ylim(273,1500)
plt.show()

Tdike = T[(T.size-1)//2]
print('Dike temp after 1 week:',Tdike,'[K]')
```

```python
T = T_naught(T)
for j in range(30*24*60*60//1800):
    T=fn(T,n)
    
plt.plot(x,T,'ro-',mfc='w')
plt.ylim(273,1500)
plt.show()

Tdike = T[(T.size-1)//2]
print('Dike temp after 30 days (1 month):',Tdike,'[K]')
```

```python
T = T_naught(T)
for j in range(365*24*60*60//1800):
    T=fn(T,n)
    
plt.plot(x,T,'ro-',mfc='w')
plt.ylim(273,1500)
plt.show()

Tdike = T[(T.size-1)//2]
print('Dike temp after 1 year:',Tdike,'[K]')
```
