```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#starting parameters
N=50
d=5 #meters
L=(d*2)+(d*2)+d
x=np.arange(0,L,.1)
T=np.zeros(5*N)
def resetT(T):
    for i in range(0,5*N):
        if x[i]>10 and x[i]<=15:
            T[i]=1474 #absolute temp
        else:
            T[i]=274   ##absolute temp
    return T
T=resetT(T)
plt.plot(x,T)
plt.ylim(0,1600)
plt.show()
```

```python
density=2.82e6  #units g/m**3
C=1.05     #units J/(g*K)
k=2.8    #units J/(s*m*K)
deltaT= 1800  #units seconds
deltaXsquared=(L/(N*5))**2 #units m**2
n=(deltaT/deltaXsquared)*(k/(C*density))
def myfunction(T,n):
    #n=.171
    Tnew=np.copy(T)
    for i in range(1,N*5-1):
        Tnew[i]=(n*((T[i+1])-(2*T[i])+T[i-1]))+(T[i])
        
    return Tnew
```

```python
T=resetT(T)
for j in range(16000):
            #T=np.copy(Tnew)
    T=myfunction(T,n)
    
#print(T[(T.size-1)//2])
plt.clf()
plt.ylim(274,1600)
plt.plot(x,T, 'ro-',mfc='w')

plt.draw()
plt.show()
plt.pause(0.05)

```

```python
T=resetT(T)
for j in range(60*60//1800):
            #T=np.copy(Tnew)
    T=myfunction(T,n)
    
print(T[(T.size-1)//2])
plt.clf()
plt.ylim(274,1600)
plt.plot(x,T, 'ro-',mfc='w')

plt.draw()
plt.show()
plt.pause(0.05)

```
