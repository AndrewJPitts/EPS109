```python
n = 11
data = np.zeros((n,n))

data[0,1] = 1
data[2,3] = 1
data[3,5] = 1
'''
These lines change individual values of the matrix in order of [row, column], but because the data begins at [0,0], data[1,1]
refers to row 2 column 2.
'''
data[5,1:n-1] = 1
data[n-2,:] = 1
'''
These lines do the same, but instead of selecting singular values of [col,row], we select many at once. 
data[5,1:n-1] refers to row 6, and columns 2 through n-1 (or 10). 
data[n-2,:] refers to row 10 and all 11 columns
'''
print(data)
plt.imshow(data,interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))

for i in range(0,n):
    data[i,3] = 1

    
'''
this for loop is saying for any i value in the a range from 0 to n (ie [0,1,2,3,...,n] where n=11),
set the values of the data = 1 in the i'th row and 3rd (4th) column for all values of i
'''    

col = (n-1) // 2
for i in range(0,n,2):
    data[i,col] = 1
    
'''
this for loop is saying for any i value in the range from 0 to n in increments of 2 (steps of 2), 
set the values of the data = 1 in the i'th row for all i, and in the column of value "col", which equals 5 (so column 6)
'''

    
print(data)
plt.imshow(data,interpolation='nearest')
plt.show()
```
