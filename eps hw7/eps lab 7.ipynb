```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 51
L = 30
step = L / N
x = np.arange(0,L,step)
```

```python
x = np.array([x])
```

```python
T = np.zeros((1,N))
T_new = np.copy(T)
T[0,0]=100
T_new[0,0]=100
```

```python
for k in range(1000):
    if k%50==0:
        plt.figure()
        plt.plot(x,T_new,'ro')
        plt.show()
    for i in range(1,N-1):
        T_new = np.copy(T)
        T_new[0,i] = 0.5*(T[0,i-1] + T[0,i+1])
        T=np.copy(T_new)
```
