```python
x_true = np.loadtxt('x_positions_2021.txt')
x_guess = np.loadtxt('Lecture_X_positions.txt')
nNumbers = len(x_guess)
```

```python
diff = x_guess - x_true

plt.figure(figsize=(13,6))
plt.plot(diff,'k')
plt.axhline(y=0,color='r',linestyle='--',label='Zero Deviation')
plt.xlabel('Guess Number',fontsize=20)
plt.ylabel('Difference Value',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.show()
```

```python
print('The most positive deviation:', diff.max())
print('The most negative deviation:', diff.min())

'''These discrepancies can be explained by errors made due to the data recording process'''
'''In certain instances, I fell behind the pace of the slides and had to catch up, making my guesses less accurate'''
```

```python
diff_avg = sum(diff)/len(diff)
print(diff_avg)

'''The sign of the average difference is positive; meaning I often over estimated the true x-value'''
```

```python
guess_avg = np.sum(x_guess)/(nNumbers)
true_avg = np.sum(x_true)/(nNumbers)

print('Average of the guessed x-values:',guess_avg)
print('Average of the true x-values:',true_avg)
```

```python
s_guess=np.array([])
s_true=np.array([])
s=np.array([])
s_av0=np.array([])
for i in range(100):
    a = ((x_guess[i] - guess_avg)**2)
    b = ((x_true[i] - true_avg)**2)
    c = ((x_true[i] - guess_avg)**2)
    d = ((x_true[i])**2)
    
    s_guess=np.append(s_guess,a)
    s_true=np.append(s_true,b)
    s=np.append(s,c)
    s_av0=np.append(s_av0,d)
    
std_guess=np.sqrt(np.sum(s_guess)/nNumbers)
std_true=np.sqrt(np.sum(s_true)/nNumbers)
std=np.sqrt(np.sum(s)/nNumbers)
print('The standard deviation of the guess x-values:',std_guess)
print('The standard deviation of the true x-values:',std_true)
print('The standard deviation between my values and the true values:',std)
```

```python
variance = std**2
print('The variance is:',variance)
```

```python
std_av0 = np.sqrt(np.sum(s_av0)/nNumbers)
variance_av0 = std_av0**2
print('The variance assuming the average is 0:',variance_av0)
```

```python
diffAll = np.array([])
for file in filenames:
    data = np.loadtxt(file);
    diffs = data - x_true
    diffAll = np.append(diffAll,diffs)

plt.hist(diffAll,50)
plt.xlim(diffAll.min(),diffAll.max()) #diffAll.min() ~ -0.3 but I excluded it for a better gaussian fit
plt.show()
```

```python
diffAll = diffAll[abs(diffAll)<0.15]
plt.hist(diffAll,28)
plt.xlim(diffAll.min(),diffAll.max())
plt.show()
```

```python
sigma = np.array([])
mean = np.array([])
std_sum_all = 0
mean_sum_all = 0

for i in range(len(diffAll)):
    std_sum_all += np.sum((diffAll[i])**2)
    mean_sum_all += np.sum(diffAll[i])
    
    std = np.sqrt(std_sum_all / len(diffAll))
    avg = mean_sum_all / len(diffAll)
    
    sigma = np.append(sigma,std)
    mean = np.append(mean,avg)

print(len(mean))
print(mean)
print(len(sigma))
print(sigma)
```

```python
from scipy.stats import norm
z = diffAll.max() - diffAll.min()
z = z/5080

x = np.arange(diffAll.min(),diffAll.max(),z)


plt.figure()
plt.hist(diffAll,25,density=True)
plt.plot(x,norm.pdf(x,mean,sigma))
plt.show()
```

### Part 3.

```python
filenames18 = glob.glob('submissions_2018/*.txt')
filenames19 = glob.glob('submissions_2019/*.txt')
filenames20 = glob.glob('submissions_2020/*.txt')
print(filenames18)
print(filenames19)
print(filenames20)
```

```python
stdDev18 = np.array([])
for file in filenames18:
    data = np.loadtxt(file);
    set_guess_avgs = np.sum(data)/(nNumbers)

    x = np.sqrt(np.sum((data - set_guess_avgs)**2)/(nNumbers-1))
        
    stdDev18 = np.append(stdDev18,x)

print(stdDev18)
print(len(stdDev18))

plt.plot(stdDev18)
plt.title('2018',fontsize=15)
plt.show()
```

```python
stdDev19 = np.array([])
for file in filenames19:
    data = np.loadtxt(file);
    set_guess_avgs = np.sum(data)/(nNumbers)

    x = np.sqrt(np.sum((data - set_guess_avgs)**2)/(nNumbers-1))
        
    stdDev19 = np.append(stdDev19,x)

print(stdDev19)
print(len(stdDev19))

plt.plot(stdDev19)
plt.title('2019',fontsize=15)
plt.show()
```

```python
stdDev20 = np.array([])
for file in filenames20:
    data = np.loadtxt(file);
    set_guess_avgs = np.sum(data)/(nNumbers)

    x = np.sqrt(np.sum((data - set_guess_avgs)**2)/(nNumbers-1))
        
    stdDev20 = np.append(stdDev20,x)

print(stdDev20)
print(len(stdDev20))

plt.plot(stdDev20)
plt.title('2020',fontsize=15)
plt.show()
```

```python
plt.figure()
plt.plot(stdDev18,label='2018')
plt.plot(stdDev19,label='2019')
plt.plot(stdDev20,label='2020')
plt.legend()
plt.show()
```

```python
std18_avg = np.average(stdDev18)
std19_avg = np.average(stdDev19)
std20_avg = np.average(stdDev20)

print('The average Standard Deviation for 2018:',std18_avg)
print('The average Standard Deviation for 2019:',std19_avg)
print('The average Standard Deviation for 2020:',std20_avg)
```

```python
'''The year of 2020 has the lowest average standard deviation which may imply that the 2020 submissions were actually
the most accurate. However, analyzing the data emprically, it can be seen that one value is significantly lower than the
rest, bringing the average below the other years. Some simple code may show this:'''

stdDev20 = stdDev20[stdDev20>0.15]
std20_avg = np.average(stdDev20)
print(std20_avg)

'''My prediction was wrong; although the 2020 std average was lowered by the aberration, it was not significant enough to explain the difference in the std between 2020 and the other years. Therefore I believe that 2020 had the most accurate guesses.'''
```
